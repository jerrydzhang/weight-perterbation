```python
import numpy as np
import matplotlib.pyplot as plt
import pysindy as ps
from models import WeightedLasso
from utils.ode import lorenz, hydrogen_bromine, hydrogen_bromine_init, map_equation, rober, rober_init
import pickle as pkl

from scipy.integrate import solve_ivp

%load_ext autoreload
%autoreload 2
```

```python
dt = 0.01
x0_train = rober_init
t_train = np.arange(0, 1e5, dt)
t_train_span = (t_train[0], t_train[-1])

data = solve_ivp(
    rober,
    t_train_span,
    x0_train,
    rtol=1e-12,
    atol=1e-12,
    method='LSODA',
    # t_eval=t_train,
)

X = data.y.T

print("Generated X shape:", X.shape)
print("t_train shape:", t_train.shape)
```

```python
for i in range(X.shape[1]):
    plt.figure(figsize=(10, 5))
    # plt.plot(t_train, X[:, i], label=f"Dimension {i+1}", lw=0.5)
    plt.plot(data.t, X[:, i], label=f"Dimension {i+1}", lw=0.5)
    plt.xlabel("Time")
    plt.ylabel(f"Trajectory Value (X{i+1})")
    plt.title(f"ODE Trajectory for Dimension {i+1}")
    plt.legend()
    plt.show()
```

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the JSON file
with open('../combined_results.json', 'r') as f:
# with open('../experiments/results/rober_results.json', 'r') as f:
    data = json.load(f)

# Process data into a DataFrame
records = []
for model, values in data.items():
    record = {'model': model}
    record.update(values['metrics'])
    record.update(values['parameters'])
    records.append(record)

df = pd.DataFrame(records)

# Replace infinity with NaN for easier processing
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Find the best model (lowest nmse_mean)
best_model = df.loc[df['nmse_mean'].idxmin()]

print("Best Model:")
print(best_model)

# Plotting
# Filter out models with NaN nmse_mean for plotting
plot_df = df.dropna(subset=['nmse_mean'])

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Parameter vs. NMSE Mean', fontsize=16)

# Parameters to plot
parameters = ['model_alpha', 'weight_eps', 'weight_factor', 'weight_std']
axes = axes.flatten()

for i, param in enumerate(parameters):
    axes[i].scatter(plot_df[param], plot_df['nmse_mean'])
    axes[i].set_xlabel(param)
    axes[i].set_ylabel('NMSE Mean')
    axes[i].set_yscale('log')  # Use log scale for NMSE Mean for better visibility
    if param != 'weight_factor':
        axes[i].set_xscale('log')  # Use log scale for better visibility
    else:
        axes[i].set_xscale('linear')  # Keep linear scale for weight_factor
        
    axes[i].set_title(f'{param} vs. NMSE Mean')
    axes[i].grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('parameter_plots.png')

# Also create individual plots for more clarity and save them
for param in parameters:
    plt.figure(figsize=(8, 6))
    plt.scatter(plot_df[param], plot_df['nmse_mean'])
    plt.xlabel(param)
    plt.ylabel('NMSE Mean')
    plt.title(f'{param} vs. NMSE Mean')
    plt.grid(True)
    plt.savefig(f'{param}_vs_nmse.png')
```

```python
from pandas.plotting import parallel_coordinates
from sklearn.preprocessing import MinMaxScaler

cols_to_plot = ['model_alpha', 'weight_eps', 'weight_factor', 'weight_std', 'nmse_mean']
data_to_scale = df[cols_to_plot]

# Use MinMaxScaler to scale each column to a range of [0, 1].
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data_to_scale)

# Create a new DataFrame with the scaled values.
df_scaled = pd.DataFrame(scaled_data, columns=cols_to_plot)

# Add the unscaled metric back to the DataFrame for coloring.
# This ensures the color bar uses the original, understandable values.
df_scaled['nmse_mean_original'] = df['nmse_mean']

# Pick out the ones where the mean is nan
df_scaled = df_scaled.dropna(subset=['nmse_mean_original'])

# --- 3. Create the Parallel Plot ---
# Set the figure size for better readability.
plt.figure(figsize=(15, 8))

# Define a colormap. 'viridis' or 'plasma' work well for this.
colormap = plt.get_cmap("plasma")

# Create the plot, using the original 'nmse_mean' to color the lines.
parallel_coordinates(
    df_scaled,
    class_column='nmse_mean_original',
    colormap=colormap,
    alpha=0.5  # Use transparency to help with overlapping lines
)

# --- 4. Add a Color Bar and Improve Readability ---
# Create a color bar to show what the colors represent.
sm = plt.cm.ScalarMappable(cmap=colormap, norm=plt.Normalize(vmin=df['nmse_mean'].min(), vmax=df['nmse_mean'].max()))
sm.set_array([]) # You need this line for the color bar to work.
cbar = plt.colorbar(sm, ax=plt.gca(), orientation='vertical')

cbar.set_label('NMSE Mean (Performance)', rotation=270, labelpad=15)

plt.title("Parallel Plot of 100 Models (Normalized)")
plt.xlabel("Parameters and Metrics")
plt.ylabel("Normalized Value")
plt.xticks(rotation=45)

# Since we are coloring by a continuous variable, a legend is not needed.
plt.legend().set_visible(False)

plt.tight_layout()
```
